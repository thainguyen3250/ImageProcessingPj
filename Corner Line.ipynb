```python
def duong_cheo_nguoc(square_area,k:int,number:int):
    row,col=square_area.shape[:2]
    if k>col:
        return
    col=col-k-1
    for i in range (row):
        if(col<0):
            break
        else:
            square_area[i][col]=number
            col=col-1
    return square_area

def corner_Line(image,start,end,thick:int):
    for k in range(thick):
        image[start:end,start:end]=duong_cheo_nguoc(image[start:end,start:end],k,0)
    return image
        
##tham số 1: ảnh        
##tham số 2: điểm bắt đầu từ tọa độ 
##tham số 3: điểm cuối 
##tham số 4: độ dày đường đen 
```

```python
plt.imshow(corner_Line(deadpool_img ,0,200,50))
```
