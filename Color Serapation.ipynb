```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
fly=cv2.imread("C:\\Users\\thainguyen5484\\Downloads\\modomdom.jpg",0)
```

```python
plt.imshow(fly,cmap="gray")
```

```python
p=25
new_image=np.copy(fly)
```

```python
for i in range(new_image.shape[0]):
    for j in range(new_image.shape[1]):
        if (new_image[i][j]>p):
            new_image[i][j]=255
        else:
            new_image[i][j]=0
    
```
